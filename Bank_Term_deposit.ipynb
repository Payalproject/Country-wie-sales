```python
job_counts = df['job'].value_counts()

print(job_counts)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='job', order=df['job'].value_counts().index)
plt.title("Job Distribution")
plt.xlabel("Count")
plt.ylabel("Job Type")
plt.show()
```

```python
# Average balance by deposit subscription

avg_balance = df.groupby('cons.price.idx')['cons.conf.idx'].mean()

print(avg_balance)
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x=avg_balance.index, y=avg_balance.values, palette="Blues_d")

plt.title("Average Balance by Deposit Subscription")
plt.xlabel("Deposit Subscription (yes/no)")
plt.ylabel("Average Balance")
plt.show()
```

```python
# Contact method analysis

count_by_method=df['contact'].value_counts()
print("Total Contacts by Method:\n", count_by_method)
```

```python
response_rate = df.groupby('contact')['poutcome'].apply(lambda x: (x == 'success').mean())
print("\n Response Rate by Method:\n", response_rate)
```
